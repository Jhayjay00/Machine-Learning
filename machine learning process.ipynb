```python
import pandas as pd

music_data = pd.read_csv('music.csv')

music_data
```

```python
x = music_data.drop(columns=['genre'])#this is our input data. We need to train the model with the values of inputs

x
```

```python
y = music_data['genre'] 

y
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()

model.fit(x.values, y)  # it's learning patterns in the data 

model.predict([[22,1], [23,0]])

```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)


model = DecisionTreeClassifier()

model.fit(x_train, y_train)  # it's learning patterns in the data 

predictions = model.predict(x_test)

score = accuracy_score(y_test, predictions)
score

```

```python
import joblib

joblib.dump(model,'music-recommender.jobliblib')
```

```python
from sklearn import tree

tree.export_graphviz(model, out_file='music-recommender.dot',
                     feature_names=['age', 'gender'],
                     class_names=sorted(y.unique()),
                     label='all',
                     rounded=True,
                     filled=True)
```
